```python
laptimes_df = pd.read_csv("chinese_gp_lap_times.csv", sep=",",)
drivers_df = pd.read_csv("chinese_gp_drivers.csv", sep=",")
result_df = pd.read_csv("chinese_gp_results.csv")

tmp = pd.merge(laptimes_df, drivers_df, left_on='Driver', right_on='Abbreviation', how="outer")

col_results = ["Abbreviation", "ClassifiedPosition", "GridPosition", "Status", "Points"]
df = pd.merge(tmp, result_df[col_results], left_on='Abbreviation', right_on='Abbreviation', how="outer")
```

```python
team_colors = {
    'Mercedes': '#27F4D2',
    'Red Bull Racing':'#3671C6',
    'Ferrari':'#E80020',
    'McLaren':'#FF8000',
    'Alpine':'#00A1E8',
    'Racing Bulls':'#6692FF',
    'Aston Martin':'#229971',
    'Williams':'#1868DB',
    'Kick Sauber':'#52E252',
    'Haas F1 Team':'#B6BABD'
}
```

```python
from matplotlib.colors import LinearSegmentedColormap
import matplotlib

# custom palette
five_palette = ['#30a2da','#fc4f30','#e5ae38','#6d904f','#8b8b8b']
cm_drivers = LinearSegmentedColormap.from_list("custom", five_palette, len(df['LastName'].unique()))


def make_palette(cm, vals:list) -> dict:
    """make a custom color map
    Args:
        matplotlib colormap, list of unique values
    Returns:
        dictionary
    """
    cols_plt_dict = {}
    for i, name in zip(range(cm.N), vals) :
        rgba = cm(i)
        cols_plt_dict[name] = matplotlib.colors.rgb2hex(rgba)
    return cols_plt_dict

drivers_palette = make_palette(cm_drivers, df['LastName'].unique())
```

```python
def convert_to_seconds(laptime:str) -> float:
    """convert a time ("00:01:45.273000") to seconds

    Args:
        laptime (str): _description_

    Returns:
        float: _description_
    """
    vals = laptime.split(":")
    return 60*int(vals[1]) + float(vals[2])

df['LapTime_seconds'] = df['LapTime'].apply(lambda x: convert_to_seconds(x))
# convert LapNumber to integer
df['LapNumber']=df['LapNumber'].astype(int)
```

```python
plt.figure(figsize=(16, 3))
sns.boxplot(data=df, x="LapNumber", y = "LapTime_seconds", showfliers=False, color="#30a2da")
plt.grid(True)
plt.show()
```

```python
teams = sorted(df['TeamName'].unique())

fig, axes = plt.subplots(len(df['TeamName'].unique()), 1, figsize=(16,10), sharex=True)
for cnt, team in enumerate(teams):
    sns.boxplot(data=df[df["TeamName"] == team], x="LapNumber", y = "LapTime_seconds",
                showfliers=False, color=team_colors[team], ax=axes[cnt])
    axes[cnt].set_ylabel("")
    if cnt != (len(teams) -1):
        axes[cnt].set_xlabel("")
    axes[cnt].grid(True)
    axes[cnt].text(.8, .8, team, transform=axes[cnt].transAxes, fontweight="bold", color=team_colors[team], fontsize=10)
```

```python
def make_joydivision_plot(tmp_df: pd.DataFrame) -> None:
    """code taken from https://seaborn.pydata.org/examples/kde_ridgeplot to make a ridge line plot
    """
    sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})
    g = sns.FacetGrid(tmp_df, row="TeamName", hue="TeamName", aspect=10, height=.5, palette=team_colors)

    # Draw the densities in a few steps
    g.map(sns.kdeplot, "LapTime_seconds",
        bw_adjust=.5, clip_on=False,
        fill=True, alpha=1, linewidth=1.5)
    g.map(sns.kdeplot, "LapTime_seconds", clip_on=False, color="w", lw=2, bw_adjust=.5)

    # passing color=None to refline() uses the hue mapping
    g.refline(y=0, linewidth=2, linestyle="-", color=None, clip_on=False)


    # Define and use a simple function to label the plot in axes coordinates
    def label(x, color, label):
        ax = plt.gca()
        ax.text(.7, .4, label, fontweight="bold", color=color, ha="right", va="center", transform=ax.transAxes)


    g.map(label, "TeamName")

    # Set the subplots to overlap
    g.figure.subplots_adjust(hspace=-.25)

    # Remove axes details that don't play well with overlap
    g.set_titles("")
    g.set(yticks=[], ylabel="")
    g.set_xlabels("")
    g.despine(bottom=True, left=True)
```
